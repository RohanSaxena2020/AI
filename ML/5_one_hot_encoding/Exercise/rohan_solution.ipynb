# Exercise

carprices.csv has car sales prices for 3 different models. First plot data points on a scatter plot chart
to see if linear regression model can be applied. If yes, then build a model that can answer
following questions,

**1) Predict price of a mercedez benz that is 4 yr old with mileage 45000**

**2) Predict price of a BMW X5 that is 7 yr old with mileage 86000**

**3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())**

```python
import pandas as pd
```

```python
df = pd.read_csv("carprices.csv")

df.shape
```

```python
df
```

```python
# plot using a scatter plot

import matplotlib.pyplot as plt

# Group the DataFrame by 'Car Model' and iterate over each group
for (car_model, group) in df.groupby('Car Model'):
    plt.scatter(group['Mileage'], group['Sell Price($)'], label=car_model)

# Labeling the plot
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.title('Scatter Plot of Cars by Model')
plt.legend()
plt.show()
```

### You'll see each car model has a clear linear coorleation, thus regression is possible

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['Car Model'], axis='columns')

final2 = final.drop(['Mercedez Benz C class'], axis='columns')

final2
```

```python
X = final2.drop('Sell Price($)', axis='columns')
X
```

```python
y = final2['Sell Price($)']

y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

### Now time for predictions

```python
# Predict price of a mercedez benz that is 4 yr old with mileage 45000

model.predict([[45000,4,0,0]])
```

### It predicts a price of ~$36991.32

```python
# Predict price of a BMW X5 that is 7 yr old with mileage 86000

model.predict([[86000,7,0,1]])
```

### It predicts a price of ~$11080.74

```python
# Tell me the score (accuracy) of your model.


model.score(X,y)
```

### Score of ~0.942

```python
#Bonus: save model using pickle

import pickle

with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```
