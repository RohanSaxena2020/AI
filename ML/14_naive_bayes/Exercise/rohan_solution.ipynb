### Naive Bayes: Exercise

Use wine dataset from sklearn.datasets to classify wines into 3 categories. Load the dataset and split it into test and train. After that train the model using Gaussian and Multinominal classifier and post which model performs better. Use the trained model to perform some predictions on test data.

```python
import pandas as pd
from sklearn import datasets
wine = datasets.load_wine()
```

```python
df = pd.DataFrame(wine.data,columns=wine.feature_names)
df['target'] = wine.target

df.head(100)
```

```python
inputs = df.drop('target',axis='columns')

inputs.head()
```

```python
targets = wine.target

targets
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'),df.target, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Now let's test Gaussian and Multinominal classifier using K fold to see which does better

```python
from sklearn.model_selection import cross_val_score

from sklearn.naive_bayes import MultinomialNB,GaussianNB
```

```python
cross_val_score(MultinomialNB(), X_train, y_train)
```

```python
cross_val_score(GaussianNB(), X_train, y_train)
```

### GaussianNB performed better

```python
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test[0:100])
```

```python
model.predict_proba(X_test[:10])
```
